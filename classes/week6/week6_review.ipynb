```python
#큐의 일반구현
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1

#isQueueFull()
def isQueueFull():
    global front, rear, queue, SIZE
    if(rear == SIZE-1):
        return True
    else:
        return False

#isQueueEmpty()
def isQueueEmpty():
    global SIZE, queue, front, rear
    if(front == rear):
        return True
    else:
        return False

#enQueue()
def enQueue(data):
    global SIZE, queue, front, rear
    if(isQueueFull()):
        print("큐가 꽉 찼어요")
        return
    rear += 1
    queue[rear] = data

#deQueue()
def deQueue():
    global SIZE, queue, front, rear
    if(isQueueEmpty()):
        print("큐가 비었어요")
        return None
    front += 1
    outData = queue[front]
    queue[front] = None
    return outData

#peek()
def peek():
    global SIZE, queue, front, rear
    if(isQueueEmpty()):
        print("큐가 비었어요")
        return None
    return queue[front+1]
```

```python
#큐의 일반 구현 완성
def isQueueFull():
    global SIZE, queue, front, rear
    if(rear == SIZE-1):
        return True
    else:
        return False

def isQueueEmpty():
    global SIZE, queue, front, rear
    if(rear == front):
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if(isQueueFull()):
        print("큐가 꽉참")
        return
    rear += 1
    queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if(isQueueEmpty()):
        print("큐가 비었음")
        return None
    front += 1
    outData = queue[front]
    queue[front] = None
    return outData

def peek():
    global SIZE, queue, front, rear
    if(isQueueEmpty()):
        print("큐가 비었음")
        return None
    return queue[front+1]
```

```python
#deQueue했으면 이동을 해줘야하는거 아님?

#큐가 꽉 안찬 조건이 rear != SIZE-1
#큐가 꽉차면? rear == SIZE-1 and front == -1
#비어있으면? rear == SIZE-1이어도 front가 -1이 아니면 비어있지 앞에

#새로운 isQueueFull()
def isQueueFull():
    global SIZE, queue, front, rear
    if(rear != SIZE-1):
        return False
    elif(rear==SIZE-1) and (front == -1): #이게 제대로 꽉찬거지 앞에도 공간이 없어
        return True
    else:
        for i in range(front+1, SIZE):
            queue[i-1] = queue[i]
            queue[i] = None
        front -= 1
        rear -= 1
        return False
```

```python
#원형 큐는 왜써?
#선형 큐는 오버헤드 발생 -> 옮겨줘도 계속 옮겨줘야되서 시간손해
#큐의 처음과 끝이 연결된 구조임

#원형 큐가 빈 경우 : rear == front
#원형 큐가 꽉 찬경우: (rear+1)%SIZE == front

def isQueueFull():
    global SIZE, queue, front, rear
    if((rear+1)%SIZE == front):
        return True
    else:
        return False

def isQueueEmpty():
    global SIZE, queue, front, rear
    if(rear==front):
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if(isQueueFull()):
        print("ㅇㄴㅇㄴ")
        return
    rear = (rear+1)%SIZE
    queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if(isQueueEmpty()):
        print("ㅇㄴㅇㄴ")
        return None
    front = (front+1)%SIZE
    outData = queue[front]
    queue[front] = None
    return outData
```
