### 맨 마지막 데이터 삭제


```python
katok = ['다현', '정연', '쯔위', '사나','지효']

def delete_data(position):

    if position < 0 or position > len(katok):
        print("데이터 삭제할 범위를 벗어남")
        return
    
    kLen = len(katok)
    katok[position] = None

    for i in range(position+1, kLen):
        katok[i-1] = katok[i]
        katok[i] = None

    del(katok[kLen-1]) #배열 맨 마지막 위치 삭제


delete_data(1)
print(katok)
```

### 문제 1

```python
def reverse_list(lst):
    n = len(lst)
    for i in range(n // 2):
        lst[i], lst[n - 1 - i] = lst[n - 1 - i], lst[i]
    return lst

example = [1, 2, 3, 4, 5]
print(reverse_list(example))  # 출력: [5, 4, 3, 2, 1]
```
